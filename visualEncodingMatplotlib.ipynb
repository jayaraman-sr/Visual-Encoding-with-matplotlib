```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#Reading the csv file
import pandas as pd
path = 'pokemon/'
all_pokemon = pd.read_csv(path + 'Pokemon.csv')

#Get the Overall attribute of all pokemon
overall = all_pokemon['Total']

#Picking the 7 most powerful pokemon(Based on Total attribute)
indices = overall.argsort()[-7:][::-1]
pokemon_stats = all_pokemon.iloc[indices, ]

#Reading all values
name = pokemon_stats['Name']
hp = pokemon_stats['HP']
attack = pokemon_stats['Attack']
defence = pokemon_stats['Defense']
special_attack = pokemon_stats['Sp. Atk']
special_defence = pokemon_stats['Sp. Def']
speed = pokemon_stats['Speed']

print(pokemon_stats)
```

# Comparing the performance of top 7 pokemon selected based on overall stats


# Bar Chart

```python
ind = np.arange(7) 
width = 0.30       
#plotting each attribute as a separate bar with width increasing in multiples
plt.bar(ind, hp, width, label='HP')
plt.bar(ind + width, attack, width,label='Attack')
plt.bar(ind + 2*width, defence, width, label='Defense')

plt.ylabel('Values')
plt.title('Pokemon Attributes')

plt.xticks(ind + width, name,rotation=90)
plt.legend(loc='best')
plt.xlabel('Pokemon')
plt.show()
```

Encodings I decided not to use, and why : Pie charts and contour plots for the Pokemon since the overall score for each of the top 7 pokemon are similar. Fot the Pie charts, the area marks are less accurate than line marks. I  also didn't choose contour plots as contours fail to show some of the heights due to limitation of vertical interval. Sections 7.4 and 8.4 from the text book. Also a stacked bar graph would have had all bars of almost the same height making it hard to interpret easily.

Advantages and disadvantages : I've taken the 3 main attributes of a pokemon, that is the Hit Points, Attack and the Defence and I've ued a simple bar plot. 
Length and color as the channels and the lines/bar as the mark makes it is easy to interpret the attributes and the values of each of the pokemon and compare them. Taken from the class lecture.
Disadvantages: Too many bars can look a bit cluttered and can be easily manipulated to give a false impression of the data.

# 3D Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in indices:
    ax.scatter(hp[i], attack[i], defence[i], marker = 'D', label = name[i])
    ax.text(hp[i], attack[i], defence[i],  name[i], zorder=1, color='k') 
    
ax.set_xlabel('HP')
ax.set_ylabel('Attack')
ax.set_zlabel('Defence')

#plt.show()

for angle in range(0, 360):
  ax.view_init(30, angle)
  plt.draw()
  plt.pause(.001)
```

P.S. The above plot is bigger and rotates when run on the terminal. In Jupyter, it seems static.

Encodings I decided not to use, and why : A stacked bar plot would have had all bars of almost the same height making it hard to interpret easily. The starting points of each of the attributes would not be alligned to a common scale complicating the calculation of vales of the attributes. Section 7.5 of the text book.

Advantages and disadvantages : Easy to focus on particular attributes when comparing and simpler to get an overall score(would be the euclidean distance from the origin).
Position and color as the channels and the points as the mark makes it is easy to interpret each of defence, attach and HP individually and compare these attributes with other pokemon. Taken from the class lecture.
Disadvantages: A little cluttered because of the annotation and needs some calculation the get the overall best pokemon.
